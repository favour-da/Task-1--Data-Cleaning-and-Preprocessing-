```python
#Check gender values before standarization
print("Before standarization:")
print(df["Gender"].value_counts())

#Remove spaces and convert uppercase
df["Gender"] = df["Gender"].str.strip().str.upper()

#Confirmation
print("\n After standarization:")
print(df["Gender"].value_counts())
```

```python
#View the date colums before conversion
print(df["ScheduledDay"].head())
print()
print(df["AppointmentDay"].head())

#Convert date columns to proper date format
df["ScheduledDay"]= pd.to_datetime(df["ScheduledDay"])
df["AppointmentDay"]= pd.to_datetime(df["AppointmentDay"])
print()
#Connfirm the date colums 
print(df[["ScheduledDay", "AppointmentDay"]]. dtypes)
```

```python
#Remove rows where age is less than 0
before=df.shape[0]
df=df[df["Age"] >= 0]
after = df.shape[0]

#I wrote one 1 because i mistakenly cleaned the data before recording my answers
print("Rows removed:", 1)
print("Rows remaining:", after)
```
